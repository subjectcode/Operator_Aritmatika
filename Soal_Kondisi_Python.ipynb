# **1. Buatlah sebuah program untuk menentukan bilangan positif, bilangan negatif, dan bilangan nol**


```python
# Inisial bilangan variabel dengan nilai awal 0
bilangan = 0

# Memeriksa apakah bilangan lebih dari 0
if bilangan > 0:
    print("Bilangan adalah lebih dari 0")
# Jika tidak periksa apakah bilangan sudah sama dengan 0
elif bilangan == 0:
    print("Bilangan adalah 0")
# Jika tidak ada kondisi sebelumnya yang terpenuhi maka bilangan dianggap kurang dari 0
else:
    print("Bilangan adalah kurang dari 0")
```

**Algoritma :**


1. Inisial variabel bilangan dengan nilai awal 0
2. Periksa apakah bilangannya lebih dari 0 Jika iya maka akan merespon "Bilangan adalah lebih dari 0"
3. Jika tidak, periksa apakah bilangan sama dengan 0 Jika iya maka akan merespon "Bilangan adalah 0"
4. Jika tidak ada kondisi sebelumnya yang memenuhi keduanya maka bilangan dianggap kurang dari 0 lalu akan merespon "Bilangan adalah kurang dari 0"

# 2. Sebuah toko sepatu menawarkan diskon 10% untuk pembelian sepatu dengan harga lebih dari Rp. 500.000 dan diskon 5% untuk pembelian sepatu lebih dari Rp. 100.000

â€¢ Buatlah program python untuk menghitung harga akhir pembelian sepatu menggunakan kondisi if, buatkan juga algoritma dan flowchartnya

```python

# Tampilkan judul program
print("-- Pembayaran Diskon Potongan Harga Sepatu --")
print("=============================================")

# Meminta pengguna untuk memasukkan nominal harga sepatu
harga_sepatu = float(input("Masukkan Harga Sepatu: "))

# Inisialisasi awal diskon dengan nilai awal 0
diskon = 0

# Memeriksa apakah harga sepatu lebih dari Rp. 500.000
if harga_sepatu > 500000:
    # Jika iya, maka berikan diskon sebesar 10%
    diskon = 0.1
    print("Hore!, Anda Mendapatkan potongan harga sekitar 10%")
# Jika tidak, periksa apakah harga sepatu lebih dari Rp. 100.000
elif harga_sepatu > 100000:
    # Jika iya, maka berikan diskon sebesar 5%
    diskon = 0.05
    print("Hore!, Anda Mendapatkan potongan harga sekitar 5%")

# Menghitung total harga setelah diskon
total_harga = harga_sepatu - (harga_sepatu * diskon)

# Menampilkan total harga setelah diskon
print("Total Harga: Rp.", total_harga)

```

**Algoritma :**

4. Proses menginialisasi diskon
5. Jika harga sepatu lebih dari Rp.500.000 maka akan diberikan diskon 10%
6. Jika harga sepatu lebih dari Rp.100.000 maka akan diberikan diskon 5%
7.Menghitung Total Harga
8. Selesai

#  3. buatlah sebuah game sederhana menggunakan bahasa python beserta algoritma dan flowchart nya  


```python
import random

# Fungsi untuk menghasilkan pilihan Musuh
def pilihan_musuh():
    return random.choice(["batu", "gunting", "kertas"])

# Fungsi untuk menentukan pemenang
def menentukan_pemenang(pilihan_pemain, pilihan_musuh):
    if pilihan_pemain == pilihan_musuh:
        return "Seri!"
    elif (pilihan_pemain == "batu" and pilihan_musuh == "gunting") or \
         (pilihan_pemain == "gunting" and pilihan_musuh == "kertas") or \
         (pilihan_pemain == "kertas" and pilihan_musuh == "batu"):
        return "Anda Menang!"
    else:
        return "Musuh Menang!"

while True:
    print("\nPermainan Jankenpon!")
    print("Pilihan Anda: (1) Batu, (2) Gunting, (3) Kertas, atau (4) Keluar")

    # Meminta pemain untuk memilih
    pilihan_pemain = input("Masukkan pilihan Anda: ")

    if pilihan_pemain == "4":
        print("Terima kasih telah bermain. Sampai jumpa!")
        break

    if pilihan_pemain not in ["1", "2", "3"]:
        print("Pilihan tidak valid. Masukkan nomor 1, 2, 3, atau 4 (untuk keluar).")
        continue

    pilihan_pemain = int(pilihan_pemain)
    if pilihan_pemain == 1:
        pilihan_pemain = "batu"
    elif pilihan_pemain == 2:
        pilihan_pemain = "gunting"
    elif pilihan_pemain == 3:
        pilihan_pemain = "kertas"

    pilihan_musuh_dua = pilihan_musuh()
    print("Musuh memilih:", pilihan_musuh_dua)

    hasil_permainan = menentukan_pemenang(pilihan_pemain, pilihan_musuh_dua)
    print("Hasil:", hasil_permainan)

```

Algoritma :

1. Mulai permainan.
2. Menginisialisasi fungsi pilihan_musuh yang akan menghasilkan pilihan acak antara "batu", "gunting", atau "kertas" untuk musuh.
3. Menginisialisasi fungsi menentukan_pemenang untuk menentukan hasil permainan berdasarkan pilihan pemain dan musuh.
4. Mulai permainan loop tak terbatas.
5. Tampilkan pesan awal permainan dan pilihan yang tersedia.
6. Meminta pemain untuk memasukkan pilihan mereka.
7. Jika pemain memilih "4", maka permainan selesai.
8. Memeriksa apakah pilihan pemain adalah salah satu dari "1", "2", atau "3".
Konversi pilihan pemain menjadi "batu", "gunting", atau "kertas" berdasarkan input yang diberikan.
9. Panggil fungsi pilihan_musuh untuk memperoleh pilihan acak musuh.
10. Tampilkan pilihan musuh.
11. Panggil fungsi menentukan_pemenang untuk menentukan hasil permainan dan tampilkan hasilnya.
12. Permainan kembali ke langkah 5 sampai pemain memilih untuk keluar.

# **4. Buatlah sebuah program sederhana menggunakan perulangan dan buatkan algoritma dan flowchartnya**

```python
# Meminta pengguna untuk memasukkan tinggi piramida
tinggi_piramida = int(input("Masukkan tinggi pyramid: "))

# Looping untuk menghasilkan piramida
for i in range(tinggi_piramida):
    for k in range(i + 1):
        print("*", end="")
    print()
```

**Algoritma :**

1. Meminta Pengguna Memasukkan tinggi piramida.
2. Inisialkan loop pertama.
3. Di dalam loop pertama, inisialkan loop kedua.
4. Cetak satu bintang(*).
5. Cetak karakter baris baru.
6. Kembali ke loop kedua.
7. Kembali ke loop pertama.
8. Selesai.
