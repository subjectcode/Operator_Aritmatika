# **Sejarah Bahasa Programan Python**
<br>

Python diciptakan pada akhir tahun 1980-an oleh Guido van Rossum di Centrum Wiskunde & Informatica (CWI) di Belanda sebagai penerus bahasa pemrograman ABC. Van Rossum mencari bahasa yang sederhana untuk dipelajari dan digunakan, namun cukup kuat untuk digunakan untuk berbagai tugas. Dia menamai bahasa tersebut Python setelah grup komedi Monty Python, yang dia sukai.

<br>

Versi pertama Python, 0.9.0, dirilis pada tahun 1991. Ini adalah bahasa kecil dengan beberapa fitur dasar, tetapi dengan cepat mendapatkan popularitas di kalangan programmer. Pada tahun 1995, Python Software Foundation (PSF) dibentuk untuk mengawasi pengembangan dan pemeliharaan bahasa tersebut.

<br>

Python 2.0 dirilis pada tahun 2000. Versi ini memperkenalkan sejumlah fitur baru, termasuk dukungan pemrograman berorientasi objek, dukungan Unicode, dan pengumpulan sampah. Python 2.0 adalah revisi besar dari bahasa tersebut, dan tidak semua kode lama kompatibel dengannya.

<br>

Python 3.0 dirilis pada tahun 2008. Versi ini merupakan perubahan yang lebih radikal dari Python 2.0, dan tidak kompatibel dengan versi sebelumnya. Python 3.0 memperkenalkan sejumlah fitur baru, seperti sintaks baru untuk string dan daftar, dan sistem tipe baru.

<br>

Saat ini, Python adalah salah satu bahasa pemrograman paling populer di dunia. Ini digunakan untuk berbagai tugas, termasuk pengembangan web, ilmu data, dan pembelajaran mesin. Python Software Foundation memperkirakan bahwa ada lebih dari 10 juta programmer Python di seluruh dunia.

<br>

Python adalah bahasa pemrograman yang kuat dan serbaguna yang digunakan oleh berbagai kalangan untuk berbagai tugas. Jika Anda mencari bahasa pemrograman yang kuat dan mudah digunakan, Python adalah pilihan yang bagus.

**Cara mengecek versi Python**

Untuk mengecek versi Python yang Anda gunakan, Anda dapat menggunakan perintah berikut di terminal atau command prompt:

```
python --version
```
atau melalui perintah
```
python3 --version
```

<br>

Perintah tersebut akan menampilkan versi Python yang Anda gunakan, beserta nama sistem operasi dan versinya.

<br>

**Cara menggunakan Pip**

Pip adalah alat manajemen paket untuk Python yang digunakan untuk menginstal dan mengelola paket perangkat lunak dari repositori Python Package Index (PyPI). Untuk menggunakan pip, Anda perlu membuka terminal atau command prompt dan mengetik perintah berikut:

```
pip install [nama_paket]
```

<br>

Misalnya, untuk menginstal paket NumPy, Anda dapat menggunakan perintah berikut:

```
pip install numpy
```

<br>

Perintah tersebut akan menginstal paket NumPy di sistem Anda.

<br>

Berikut adalah beberapa perintah pip yang umum digunakan:

```
pip install [nama_paket]: Menginstal paket perangkat lunak.
```
```
pip uninstall [nama_paket]: Menghapus paket perangkat lunak.
```
```
pip freeze: Menampilkan daftar paket yang diinstal.
```
```
pip list: Menampilkan daftar paket yang diinstal beserta versinya.
```
```
pip show [nama_paket]: Menampilkan informasi tentang paket perangkat lunak.
```
